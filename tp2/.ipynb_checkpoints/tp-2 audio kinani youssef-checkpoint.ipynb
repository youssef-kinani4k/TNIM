```python
import wave
import matplotlib.pyplot as plt
import numpy as np
import sounddevice as sd
import soundfile as sf
import tkinter as tk
from tkinter import filedialog as fd
from pydub import AudioSegment
```

```python
#Exercice 1: Signal Audio

#1) charger un fichier
def  nom_fichier():
    fenetre=tk.Tk()
    n=fd.askopenfilename()
    fenetre.destroy()
    return n
def audio_open(audio_name):
    file=wave.open(audio_name,"rb")
    return file

#2) determiner le nombre de pistes
def nbr_pistes(audio):
    nch=info[0]
    if nch==1:
        print("Il s'agit d'un son mono")
    elif nch==2:
        print("Il s'agit d'un son stereo")
    elif nch==3:
        print("Il s'agit d'un son dolpy")

#3) fonction qui extraire chaque piste audio dans un vecteur à part.
def lire_audio(audio):
    audio.rewind()
    audio_data=audio.readframes(info[3])
    audio_data=np.frombuffer(audio_data,dtype=np.int16)
    return audio_data
def separer_piste(audio):
    audio_str=lire_audio(audio)
    if(info[0]==1):return audio_str
    PS=[]
    for i in range(info[0]):
        piste=audio_str[i::info[0]]
        PS.append(piste)
    return PS
#4) représenter graphiquement la composante du singal.
def representer(audio):
    d=separer_piste(audio)
    if(info[0]==1):plt.plot(d)
    else:
        for i in range(info[0]):
            plt.subplot(2,1,i+1)
            plt.plot(d[i])

#5) fonction qui permet d'extraire une partie d'une piste délimitée par les indices K1 et K2.
def extraire_partie(audio,k1,k2,canal):#k1 et k2 en seconde
    info=audio.getparams()
    d=separer_piste(audio)
    if canal > info[0]:
        print("canal eror")
        return -4
    ext=d[canal-1][k1*info[2]:k2*info[2]]
    return ext
#6) fonction qui permet de representer graphiquement en fonction du temps la partie extraite de la piste.
def presenter_temps(d):
    Te=1/(info[2])
    t=[Te*i for i in range(len(d))]
    plt.plot(t,d)

#7) ecouter un a un  
def ecouter_un_A_un(audio):
    s=lire_audio(audio)
    a=separer_piste(audio)
    sd.play(a[0][0:len(a[1])//20],info[2]*1,1)
    sd.wait()
    sd.play(a[1][len(a[1])//20:len(a[1])//10],info[2]*1,2)
    sd.wait()
    sd.play(s[len(s)//10:len(s)//5],info[2]*2)
    sd.wait()
    sd.stop()

#8) fonction qui permet générer les fichiers .wav relatifs à chaque pqrtie extraite.
def save_chaque_piste(audio):
    s=separer_piste(audio)
    for i in range(info[0]):
        p=wave.open("p"+str(i+1)+".wav","wb")
        p.setparams(info)
        p.setnchannels(1)
        p.writeframes(s[i].tobytes())
#9) fonction qui permet de réaliser un fichier audio ou on peut entendre en alternance, 
    #le son séparement sur chaqun des baffes et les 2 simultanéments
def droit_gauche(audio,s):#prend list de 3 valeur
    d=lire_audio(audio)
    l=np.zeros_like(d)
    l[:s[0]*info[2]*2:2]=d[:s[0]*info[2]*2:2]
    l[s[0]*info[2]*2+1:s[1]*info[2]*2:2]=d[s[0]*info[2]*2+1:s[1]*info[2]*2:2]
    l[s[1]*info[2]*2:s[2]*info[2]*2:1]=d[s[1]*info[2]*2:s[2]*info[2]*2:1]
    w=wave.open("droit_gauche_simultanéments.wav","wb")
    w.setparams(audio.getparams())
    w.writeframes(l[:s[2]*info[2]*2].tobytes())
#Exercoce 2: Echantillonnage et codage
#1) fonction qui permet d'extraire le maximum d'information sur un fichier audio lu.
def max_info(audio):
    param=audio.getparams()
    print("le nombre de pistes:",param[0])
    print("le nombre des echantillons: ",param[3])
    print("le nombre de bits par echantillon: ",param[1]*8)
    print("la frequence d'echantillonnage: ",param[2],"Hz")
    print("le type de compression: ",param[4])
    print("nom de compression: ",param[5])
    print("la duree en seconde: {: .2f}".format(param[3]/param[2]))
    print("la taille en Ko: {: .2f}".format((param[0]*param[1]*param[3])/1024))
#2) fonction qui permet d'augmenter/dimunier progressivement l'amplitude du son et de l'ecouter.
def augmenter(audio):
    d=lire_audio(audio)
    dc=[]
    dc[:len(d)//40]=d[:len(d)//40]*1.5
    dc[len(d)//40:len(d)//15]=d[len(d)//40:len(d)//15]*0.2
    dc[len(d)//15:len(d)//6]=d[len(d)//15:len(d)//6]
    dc=np.array(dc,dtype=np.int16)
    dc=np.clip(dc,-32000,32000)
    w=wave.open("augmenter_dimunier volume.wav","wb")
    w.setparams(audio.getparams())
    w.writeframes(dc.tobytes())
#3) 
#a-sous_échantillonnage.
def echan(audio,y):
    l=lire_audio(audio)
    t=[]
    for i in range(0,len(l),int(y*2)):
        t.append(l[i])
        t.append(l[i+1])
    return np.array(t,dtype=np.int16)
#b et c
def tracer_ecouter_freqMax(audio,y):
    l=lire_audio(audio)
    ech=echan(audio,y)
    plt.subplot(2,1,1)
    plt.plot(l)
    plt.subplot(2,1,2)
    plt.plot(ech)
    sd.play(l[:2*info[2]*2],info[2]*2)
    sd.wait()
    sd.play(ech[:2*info[2]*2//y],info[2]*2/y)
    sd.wait()
    fEch=info[2]/y
    print("taille= ",info[1]*len(ech)/(1024**2),"frequence maximal= ",fEch*2)
    
#4) requantification.
def requantifier(audio,r):  
    d=audio.raw_data
    i=audio.sample_width
    i=int(i*8)
    data1=np.frombuffer(d,dtype="int"+str(i))
    da=data1*((2**r)/(2**i))
    da=np.array(da,dtype="int"+str(r))
    da=da.tobytes()
    sound=AudioSegment(data=da , sample_width=int(r*(1/8)) , frame_rate=audio.frame_rate ,channels=audio.channels)
    print("taille initial = ",len(data)*audio.sample_width,"oct"," ||  taille final = ",len(data)*sound.sample_width,"oct")
    fp=sound.export("requantifier"+str(r)+".mp3")
#Exercice 3: Compression de signaux audio et évaluation du taux de compression.
def compresser(file,forma):
    from pydub import AudioSegment
    from os.path import splitext,getsize
    f=AudioSegment.from_file(file)
    x=splitext(file)[0]+"_"+splitext(file)[1][1:]+"to"+forma+"."+forma
    f.export(x,format=forma)
    print("le taux de compression : {:.2f}".format(getsize(x)/getsize(file)))
    print(x)
    
#6) fonction qui permet de lire tous les fichiers audio existant dans un répertoire.
def lire(folder_path):
    import os
    from pydub import AudioSegment
    from pydub.playback import play
    for file in os.scandir(folder_path):
        print(file)
        song = AudioSegment.from_file(file)
        play(song)
#7) interface graphique
from functools import partial
def compress_button_clicked(format_entry):
    input_path = nom_fichier()
    if input_path:
        output_format = format_entry.get()
        compressed_file_path = compresser(input_path, output_format)
        print(f"Fichier compressé sauvegardé sous : {compressed_file_path}")

def create_ui():
    root = tk.Tk()
    root.title("Audio Compression")

    # Définir la taille de la fenêtre (largeur x hauteur)
    root.geometry("300x300")

    format_label = tk.Label(root, text="Catégorie de compression:")
    format_label.pack()

    categories = [("avec perte", ["mp3", "ogg","aac","wma"]), (" sans perte", ["wav", "flac","aiff","alac"])]  
    format_var = tk.StringVar(root)

    for category, formats in categories:
        category_label = tk.Label(root, text=category)
        category_label.pack()

        for format in formats:
            tk.Radiobutton(root, text=format, variable=format_var, value=format).pack()

    compress_button = tk.Button(root, text="Choisir et Compresser", command=partial(compress_button_clicked, format_var))
    compress_button.pack()

    root.mainloop()
```
